```python
# create a VBox to hold all of the elements to display the particle tab
particle_box = ipywidgets.widgets.VBox([
        ipywidgets.widgets.HTML(
'''
<h1>Magnetic Force on Charged Particle</h1>
Using magnetic force right-hand-rule, select the direction 
of the magnetic force on the charged particle.'''),
        magneticrhr.MagneticRHR('particle')])
```

```python
# create a VBox to hold all of the elements to display the wire tab
wire_box = ipywidgets.widgets.VBox([
        ipywidgets.widgets.HTML(
'''
<h1>Magnetic Force on a Wire</h1>
Using magnetic force right-hand-rule, select the direction 
of the magnetic force on the current-carrying wire.
'''),
        magneticrhr.MagneticRHR('wire')])
```

```python
tab = ipywidgets.widgets.Tab(children=[
    particle_box, 
    wire_box])
tab.set_title(0, 'Particle')
tab.set_title(1,'Wire')
tab
```
